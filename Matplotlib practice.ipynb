```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure(figsize=(10,4))
a = list(range(1,11))
b = [i**2 for i in a]
plt.title("LinePlot")
plt.xlabel("a_values")
plt.ylabel("b_values")
plt.plot(a,b,color= "r",marker="x",label = "a-b Plot")
plt.legend(fontsize = "xx-large",loc=4)
plt.show()
```

```python
plt.figure(figsize=(5,5))
a = np.random.randint(1,100,50)
b = np.random.randint(1,100,50)

plt.scatter(a,b,color="black",marker = "*",s = 60)
plt.show()
```

```python
x = np.arange(0,4*np.pi,0.1)   # start,stop,step
a = np.sin(x)
b = np.cos(x)
c = np.tan(x)

plt.plot(x,a,color="r",label = "sin_plot")
plt.plot(x,b,color="g",label = "cos_plot",linestyle = (5,(10,3)))
plt.plot(x,c,color="black",label = "tan_plot",linestyle=(0,(1,1)))
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(8,4))
con = ['India',"China","US","Brazil","Russia"]
pop = [1450935791,1419321278,345426571,211998573,144820423]

plt.bar(con,pop,color=["g","r","b",'black','y'])
plt.show()
```

```python
plt.figure(figsize=(8,4))
con = ['India',"China","US","Brazil","Russia"]
pop = [1450935791,1419321278,345426571,211998573,144820423]

plt.pie(pop,labels=con,autopct="%.2f",explode=[0,0,0.1,0,0.2])
plt.show()
```

```python
plt.figure(figsize=(8,4))
com = ["RI","TCS","HDFC","Airtel","ICICI"]
cap = [2687,4050,1714,1691,1267]
plt.title("Company-Cap")

plt.pie(cap,labels = com,autopct="%.2f",explode=[0,0.1,0,0,0.1],colors=['r','b','g','black','y'])
plt.legend(fontsize="x-small")
plt.show()
```

```python
a = np.random.randint(1,1000,500)
plt.title("Histogram")
plt.hist(a,bins=30,color= "cyan",label=[str(i) for i in range(1,31)])
plt.show()
```

```python
plt.figure(figsize=(16,4))
a = list(range(1,11))
b = [i**2 for i in a]
c = np.random.randint(1,20,10)
d = np.random.randint(1,30,10)
con = ['India',"China","US","Brazil","Russia"]
pop = [145,141,34,21,14]

plt.subplot(1,4,1)
plt.title("LinePlot")
plt.plot(a,b,color="cyan",label="L-plot")
plt.legend()
plt.subplot(1,4,2)
plt.title("ScatterPlot")
plt.scatter(c,d,color="r",label ="S-plot")
plt.legend()
plt.subplot(1,4,3)
plt.title("BarPlot")
plt.bar(con,pop,color=['r','black','g','y','b'],label=con)
plt.legend()
plt.subplot(1,4,4)
plt.title("HistPlot")
plt.hist(c,bins= 8,color="m",label="H-plot")
plt.legend()
plt.show()
```

```python
a = np.random.randint(1,20,100)
b = a.reshape(10,10)
plt.imshow(b,cmap="plasma",interpolation='nearest')
plt.colorbar()
plt.show()
```

```python
a = np.random.randint(1,10,5)
b = [i**2 for i in a]
plt.xticks([])
plt.yticks([])
plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")
plt.plot(a,b,color="m",label = "a-b plot")
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.show()
```
