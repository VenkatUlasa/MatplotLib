matplotlib is a widely used Python library for creating static, interactive, and animated visualizations. It provides an interface for creating a wide variety of plots such as line charts, bar charts, histograms, scatter plots, and more. It's highly customizable and often used in conjunction with other libraries like NumPy and pandas.


=> Figure: The entire figure or window containing all plot elements.

=> Axes: A single plot or graph within a figure. Multiple axes can exist within a single figure.

=> Axis: Refers to the x-axis or y-axis in a plot.

=> Plot: The graphical representation of data (such as a line, bar, or scatter plot).










**plt.plot(): For line plots.

**plt.bar(): For bar charts.

**plt.scatter(): For scatter plots.

**plt.hist(): For histograms.

**plt.title(), plt.xlabel(), plt.ylabel(): For adding text elements.

**plt.legend(): To show legends.

**plt.savefig(): To save a figure.

**plt.subplots(): For creating multiple plots (subplots).

```python
import matplotlib.pyplot as plt
```

### Scatter Plot

it is a plot with scatters and it works with 2D data only.

```python
a = [2,5,1,8,6,9]
b = [5,3,2,4,9,7]

plt.scatter(a,b)
plt.show()
```

```python
# changing figure sizes(length and width);
plt.figure(figsize=(8,3))
a = [2,5,1,8,6,9]
b = [5,3,2,4,9,7]
plt.title("Scatter_Plot") # titling the figure
plt.scatter(a,b)
plt.show()
```

```python
# labeling x and y axis ;
plt.figure(figsize=(8,3))
a = [2,5,1,8,6,9]
b = [5,3,2,4,9,7]
plt.title("Scatter_Plot") # titling the figure
plt.xlabel("a-Values")
plt.ylabel("b-Values")
plt.scatter(a,b)
plt.show()
```

```python
# Modifing Scatters ;
plt.figure(figsize=(8,3))
a = [2,5,1,8,6,9]
b = [5,3,2,4,9,7]
plt.title("Scatter_Plot") # titling the figure
plt.xlabel("a-Values")
plt.ylabel("b-Values")
plt.scatter(a,b,color = "r",marker = "*")  #
plt.show()
```

```python
# Hiding values from x and y axis using xticks()  and  yticks() :
plt.figure(figsize=(8,3))
a = [2,5,1,8,6,9]
b = [5,3,2,4,9,7]
plt.title("Scatter_Plot") # titling the figure
plt.xlabel("a-Values")
plt.ylabel("b-Values")
plt.xticks([])
plt.yticks([])
plt.scatter(a,b,color = "r",marker = "*")  #
plt.show()
```

### LinePlot

in this type the points are connected with a line .

** plt.plot()

```python
plt.figure(figsize=(8,3))
a = [2,5,1,8,6,9]
b = [5,3,2,4,9,7]
plt.title("Line_Plot") # titling the figure
plt.xlabel("a-Values")
plt.ylabel("b-Values")
plt.plot(a,b,color = "r",marker = "*")  #
plt.show()
```

### Legend()

```python
# if we have multiple plots in a figure;
plt.figure(figsize=(8,3))
a = [2,5,1,8,6,9]
b = [5,3,2,4,9,7]
c = [3,6,8,5,2,4]
plt.title("Scatter_Plot") # titling the figure
plt.xlabel("a-Values")
plt.ylabel("b-Values")
plt.plot(a,b,color = "r",marker = "*",label = "a-b plot")  #
plt.plot(a,c,color = "g",marker = "X",label = "a-c plot")
plt.legend()  # we can modify the legend by giving parameters;
plt.show()
```

```python
# we can modify legend using parameters ;
plt.figure(figsize=(8,3))
a = [2,5,1,8,6,9]
b = [5,3,2,4,9,7]
plt.title("Scatter_Plot") # titling the figure
plt.xlabel("a-Values")
plt.ylabel("b-Values")
plt.plot(a,b,color = "r",marker = "*",label = "a_b_plot")  #
plt.legend(loc=2,fontsize = "xx-large")  # loc = 2 means location is upper left .
plt.show()
```

### Subplot()

it is the function used to divide the figure having multiple plots into sub plots.

in subplot(1,2,1) => 1 is no of rows ; 2 is no of columns  ; 1 is position of plot.

```python
# subplots horizontally
plt.figure(figsize=(8,3))
a = [2,5,1,8,6,9]
b = [5,3,2,4,9,7]
c = [3,6,8,5,2,4]
plt.title("LinePlot") # titling the figure

plt.subplot(1,2,1)
plt.plot(a,b,color = "r",marker = "*",label = "a-b plot")  #
plt.legend()
plt.subplot(1,2,2)
plt.plot(a,c,color = "g",marker = "X",label = "a-c plot")
plt.legend()  # we can modify the legend by giving parameters;
plt.show()
```

```python
# subplots horizontally
plt.figure(figsize=(8,3))
a = [2,5,1,8,6,9]
b = [5,3,2,4,9,7]
c = [3,6,8,5,2,4]
plt.title("Scatter_Plot") # titling the figure

plt.subplot(2,1,1)
plt.plot(a,b,color = "r",marker = "*",label = "a-b plot")  #
plt.legend()
plt.subplot(2,1,2)
plt.plot(a,c,color = "g",marker = "X",label = "a-c plot")
plt.legend()  # we can modify the legend by giving parameters;
plt.show()
```

### Histogram

A histogram is a type of plot that allows you to visualize the distribution of a dataset by dividing the data into bins and showing the frequency (count) of data points in each bin. It is commonly used to understand the underlying frequency distribution of data points.

```python
plt.figure(figsize=(8,4))

a = [4,6,3,10,12,17,15,14,19,22,25,21,16,26,35,33,39,40,32]

plt.hist(a,bins=4,color="black")
plt.show()
```

### Bar Plot

A bar plot (or bar chart) is a type of graph used to represent categorical data with rectangular bars. The height (or length) of each bar corresponds to the value it represents. Bar plots are useful for comparing different categories or showing the frequency of data in different groups.

=>  plt.bar(): Creates vertical bar plots.

=>  plt.barh(): Creates horizontal bar plots.

=>  plt.xticks(): Customizes the tick marks on the x-axis.

=>  plt.text(): Adds data labels to each bar.

=>  width: Controls the width of the bars.

```python
plt.figure(figsize=(8,4))

a = ["Google","Meta","TCS","IBM"]
b = [1500,1000,1300,800]

plt.bar(a,b,color= ["black",'g','r','b'])
plt.show()
```

```python
plt.figure(figsize=(8,4))

a = ["Google","Meta","TCS","IBM"]
b = [1500,1000,1300,800]

plt.bar(a,b,color= ["black",'g','r','b']) # by default all the bars will be in same color;
plt.show()
```

### Pie Plot


A pie plot (or pie chart) is used to display proportions of categories within a whole. Each wedge of the pie represents a category, and the size of the wedge corresponds to its value relative to the total.


```python
plt.figure(figsize=(8,4))

a = ["Google","Meta","TCS","IBM"]
b = [1500,1000,1300,800]

plt.pie(b,labels=a)
plt.show()
```

```python
plt.figure(figsize=(8,4))

a = ["Google","Meta","TCS","IBM"]
b = [1500,1000,1300,800]

plt.pie(b,labels=a,autopct="%.2f")
plt.show()
```

```python
plt.figure(figsize=(8,4))

a = ["Google","Meta","TCS","IBM"]
b = [1500,1000,1300,800]

plt.pie(b,labels=a,explode=[0.1,0,0.2,0],autopct="%.2f")
plt.show()
```
